```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# credit-g
model = ["OneR", "AdaBoostM1", "NB", "LR", "J48", "RF"]
estimates = ["Accuarcy", "AUC", "RMSE"]
acc = [0.663, 0.709, 0.753, 0.757, 0.709, 0.733]
auc = [0.5288, 0.7254, 0.7815, 0.7888, 0.6398, 0.738]
rmse = [0.5805, 0.4292, 0.4228, 0.4048, 0.4789, 0.4248]
```

```python
plt.plot(np.arange(6), acc, color = "r", marker = "o")
plt.plot(np.arange(6), auc, color = "b", marker = "+")
plt.plot(np.arange(6), rmse, color = "g", marker = "*")
plt.legend(estimates)
scale = range(6)
plt.xticks(scale, model)
plt.show()
```
